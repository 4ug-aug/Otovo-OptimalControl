```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p_value_og   = pd.read_csv("data/p_value_og.csv")
p_value_sqrt = pd.read_csv("data/p_value_sqrt.csv")
p_value_cube = pd.read_csv("data/p_value_cube.csv")

trains_og    = pd.read_csv("data/trains_og.csv")
trains_sqrt  = pd.read_csv("data/trains_sqrt.csv")
trains_cube  = pd.read_csv("data/trains_cube.csv")

folds_og     = pd.read_csv("data/folds_og.csv")
folds_sqrt   = pd.read_csv("data/folds_sqrt.csv")
folds_cube   = pd.read_csv("data/folds_cube.csv")

p_value_og   = p_value_og.to_numpy()
p_value_sqrt = p_value_sqrt.to_numpy()
p_value_cube = p_value_cube.to_numpy()

print(p_value_og.shape)

trains_og   = trains_og.to_numpy().reshape(-1,10)
trains_sqrt = trains_sqrt.to_numpy().reshape(-1,10)
trains_cube = trains_cube.to_numpy().reshape(-1,10)

folds_og    = folds_og.to_numpy()
folds_sqrt  = folds_sqrt.to_numpy()
folds_cube  = folds_cube.to_numpy()
```

```python
plt.hist(trains_og)
plt.title("Histogram of unchanged SARIMA training set values")
plt.savefig("plots/"+"Histogram of unchanged SARIMA training set values"+".pdf",format="pdf",transparent=False, 
                    edgecolor="none",bbox_inches='tight',dpi=250)
plt.show()
print(folds_og.flatten())

plt.hist(trains_sqrt)
plt.title("Histogram of square-rooted SARIMA training set values")
plt.savefig("plots/"+"Histogram of square-rooted SARIMA training set values"+".pdf",format="pdf",transparent=False, 
                    edgecolor="none",bbox_inches='tight',dpi=250)
plt.show()
print(folds_sqrt.flatten())

plt.hist(trains_cube)
plt.title("Histogram of cube-rooted SARIMA training set values")
plt.savefig("plots/"+"Histogram of cube-rooted SARIMA training set values"+".pdf",format="pdf",transparent=False, 
                    edgecolor="none",bbox_inches='tight',dpi=250)
plt.show()
print(folds_cube.flatten())
```

```python
p_values = np.array([p_value_og,p_value_sqrt,p_value_cube]).reshape(3,10).transpose()
plt.boxplot(p_values, )
plt.title("Boxplots of p-values")
plt.xticks(np.arange(3)+1, ['Unchanged', 'Square-rooted', 'Cube-rooted'])
plt.savefig("plots/"+"Boxplots of p-values"+".pdf",format="pdf",transparent=False, 
                    edgecolor="none",bbox_inches='tight',dpi=250)
plt.show()
```
